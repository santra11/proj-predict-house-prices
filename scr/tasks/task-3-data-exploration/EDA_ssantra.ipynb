### Description of the data

- indType: This column contain information about the type  of the property.

- valMonthIptu : They may be monetary values related to property taxes and condominium fees.

### Choosing the threshold at 80 percent

```python
def drop_columns_with_high_missing_values(df, threshold=0.80):
    total_rows = len(df)
    for col in df.columns:
        missing_count = df[col].isnull().sum()
        if missing_count / total_rows > threshold:
            df.drop(col, axis=1, inplace=True)
    return df
```

```python
df.columns
```

```python
# Original column names
original_columns = df.columns.tolist()

# Call the drop_columns function
df = drop_columns_with_high_missing_values(df)

# Column names after dropping
remaining_columns = df.columns.tolist()

# Find the dropped columns
dropped_columns = [col for col in original_columns if col not in remaining_columns]

# Print the dropped columns
print("Dropped columns:", dropped_columns)
```

### Skewed Data columns

```python
# Plot each column and identifying skewed distributions
skewed_columns = []

for column in df.columns:
    if df[column].dtype in [int, float]:
        plt.figure(figsize=(8, 6))
        sns.kdeplot(df[column])
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
        
        if df[column].mean() != df[column].median():
            skewed_columns.append(column)

print(skewed_columns)
```

```python
len(skewed_columns)
```

```python
#Remove the NaN values
# Calculate medians for the selected columns
medians = df[skewed_columns].median()

# Impute missing values with medians
df[skewed_columns] = df[skewed_columns].fillna(medians)
```

```python
# Impute the categorical data with mode
mode_value = df['namCondominium'].mode()[0]

# Impute missing values in the selected column with the mode
df['namCondominium'].fillna(mode_value, inplace=True)
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# We can drop more columns according to high correlation

# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features from the train data
df=df.drop(columns=to_drop)
```

```python
# Get rid of outliers using capping
# Define lower and upper bounds (e.g., based on percentiles)
lower_bound = df['valSales'].quantile(0.05)  # Set to the 5th percentile
upper_bound = df['valSales'].quantile(0.95)  # Set to the 95th percentile


df['valSales'] = np.where(df['valSales'] >= upper_bound, upper_bound, 
                          np.where(df['valSales'] <= lower_bound,lower_bound,df['valSales']))
```
